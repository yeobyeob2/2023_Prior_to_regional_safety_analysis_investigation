```python
import folium
from folium import plugins
import pandas as pd
import json  
```

```python
df = pd.read_csv('/Users/isang-yeob/Downloads/KP2020.csv', encoding='cp949') #데이터 불러오기
```

```python
vp_df = df.query('EVT_CL_CD == 215') #보이스피싱 사건만 필터링
```

```python
vp_df
```

```python
hppn_xy = vp_df[['HPPN_X','HPPN_Y']] #발생지점만 추출하여 새로운 df 생성
```

```python
hppn_xy_dr_na = hppn_xy.dropna() #결측값 삭제
```

```python
hppn_yx = hppn_xy_dr_na[['HPPN_Y', 'HPPN_X']] #X좌표 Y좌표 순서 변경
```

```python
hppn_yx
```

```python
m = folium.Map(location=[36.5184, 127.3845475],  zoom_start=9) 
```

```python
N = len(hppn_yx)
```

```python
popups = [str(i) for i in range(N)]
```

```python
plugins.MarkerCluster(hppn_yx, popups=popups).add_to(m) #지도상에 보이스피싱 사고건수에 대한 마커 표시
```
